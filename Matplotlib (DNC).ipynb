### Histogram 
(Quantitative variables). For a continuous data.

```python
ages = [70,66,40,30,32,44,90,20,9,18,8,22,56,33,67,2,36,19,45,33,35,65,4,27,17] 
bins = 10                        # setting the no. of intervals.  
plt.hist(ages, bins, color = 'green')# plotting a histogram   
plt.xlabel('age')                # x-axis label
plt.ylabel('No. of people')      # frequency label 
plt.title('My histogram')        # plot title 
plt.show()                       # function to show the plot 
```

### Bar diagram 
(Catagorical variables)

### Pie Chart

```python
persons = [40,26,33,70]
activities = ['sleeping','eating','playing','running']
cols = ['c','m','r','b']
plt.pie(persons,labels = activities,
        colors = cols, explode = explode,
        shadow = True,
        startangle = 90, # starting angle of 90 degree on sleeping &rotates counterclockwise.
        autopct = '%1.1f%%')# put percentage on the slides
plt.title("pie plot",fontweight = 'bold',fontsize = '20', color = 'r')
explode = (0, 0.4, 0, 0)
plt.legend(title = "activities:", loc = 4)
plt.show()
```

### Infinite square well potential 
Solution of Schrodinger's equation is
$$\psi(x)= \sqrt{2\over a}sin{n\pi x\over a}$$
Energy $E = {{n^2h^2\pi^2}\over {2ma^2}}$

```python
pi = np.pi
a = 2*pi
xvals = np.linspace(0,a,1000)# Evaluating wavefunction at these x values. 
A = (2/a)**(1/2)
def psi(n,x):
    wf = A*np.sin(n*pi*x/a)
    return wf
yvals1 = psi(1,xvals)
yvals2 = psi(2,xvals)
yvals3 = psi(3,xvals)
plt.plot(xvals,yvals1,'m',linewidth = 2.0, label = "n = 1") # always in +ve energy state.
plt.plot(xvals,yvals2,'b',linewidth = 1.0, label = "n = 2")
plt.plot(xvals,yvals3,'g',linewidth = 1.0, label = "n = 3")
plt.xlabel("X-values",color = 'r', fontweight = 'bold',fontsize = '10')
plt.ylabel("$\psi$ (x)",color = 'r', fontweight = 'bold',fontsize = '10')
plt.title("Infinite square well potential")
plt.legend()
plt.grid()
plt.text(0,-0.08,'o',color = 'r', fontweight = 'bold',fontsize = '13')
plt.text(6.3,-0.08,'a',color = 'r', fontweight = 'bold',fontsize = '13')
plt.show()
```

Gaussian Distribution

```python
mean = 0; std = 1; variance = np.square(std)
x = np.arange(-5,5,.01)
f = np.exp(-np.square(x-mean)/2*variance)/(np.sqrt(2*np.pi*variance))
plt.plot(x,f)
plt.ylabel('Gaussian Distribution')
plt.xlabel('X-values')
plt.show()
```
